## Traversing up Azimuth level 3

### Tree traversal for matching Az to Asctb

```python
def check_in_asctb_1(cl_az,i,all_links_az):    
    flag=0
    for j in range(len(asctb_kidney_all_cts_label_unique['CT/ID'])):
        if cl_az == asctb_kidney_all_cts_label_unique['CT/ID'][j]:
            tree_match_asctb.append(j)
            tree_match_az.append(i)
            flag=1
            hierarchy_list.append([[hierarchy],[i],[j]])
            print(cl_az,asctb_kidney_all_cts_label_unique['CT/ID'][j],"Match found")
            
    if flag==0:
        print(cl_az)
        ols_call(cl_az,i,all_links_az)
```

```python
def ols_call(cl_az,i,all_links_az):
    #t=0
    url = "http://www.ebi.ac.uk/ols/api/ontologies/cl/terms?iri=http%3A%2F%2Fpurl.obolibrary.org%2Fobo%2F"
    payload={}
    headers = {
      'Accept': 'application/json'
    }
    #Azimuth
    try:
        response = requests.request("GET", all_links_az['parents']['href'], headers=headers, data=payload)
    except:
        print("No parent")
        tree_not_match.append(i)
        hierarchy_list.append([[hierarchy],[i]])
        return

    if response.status_code!=200:
        print("Status !=200")
        tree_not_match.append(i)
        hierarchy_list.append([[hierarchy],[i]])

    else:
        result_az= json.loads(response.text)
        all_links_az=result_az['_embedded']['terms'][0]['_links']
        ct_id_az=result_az['_embedded']['terms'][0]['obo_id']
        label_az=result_az['_embedded']['terms'][0]['label']
        hierarchy[ct_id_az]=label_az
        if ct_id_az[:-8:-1]=='0000000':
            hierarchy[ct_id_az]= label_az
            tree_not_match.append(i)
            hierarchy_list.append([[hierarchy],[i]])
            print(ct_id_az, "No match")
        else:
            hierarchy[ct_id_az]= label_az

            check_in_asctb_1(ct_id_az,i,all_links_az)
```

```python
tree_match_asctb=[]
tree_match_az=[]
tree_not_match=[]
hierarchy_list=[]
# tree_match_ct=[]
# tree_match_label=[]

url = "http://www.ebi.ac.uk/ols/api/ontologies/cl/terms?iri=http%3A%2F%2Fpurl.obolibrary.org%2Fobo%2F"
payload={}
#hierarchy={}
headers = {
  'Accept': 'application/json'
}
for i in range(len(az_mismatch_asctb['CT/ID'])):
    hierarchy={}
    cl_az=az_mismatch_asctb['CT/ID'][i]
    print(cl_az,"Original")
    hierarchy[cl_az]=az_mismatch_asctb['CT/LABEL'][i]
    cl_az=cl_az.replace(":","_")
    response = requests.request("GET", url+cl_az, headers=headers, data=payload)
    if response.status_code!=200:
        tree_not_match.append(i)
        hierarchy_list.append([hierarchy,i])
    else:
        result_az= json.loads(response.text)
        all_links_az=result_az['_embedded']['terms'][0]['_links']
        ols_call(cl_az,i,all_links_az)    
```
